```python
# filtering revenue to premium subscription

for df in [ptg, pcg1, pcg2]:
    df.loc[df['product_id'].isnull(), 'product_id'] = ''
    df.loc[:, 'has_changed'] = 0
    df.loc[df['payment_provider'].isin(['Деньги от Payment Ninja', 'Деньги от Google Play']), 'has_changed'] = 1
    df.loc[~df['product_id'].str.contains('vip|prem', regex=True), 'has_changed'] = 0  
    df.loc[:, 'test_revenue'] = 0
    df.loc[df['has_changed'] == 1, 'test_revenue'] = df.loc[df['has_changed'] == 1, 'revenue']
```

```python
# calculating and merging revenue per user

r_ptg = ptg.groupby('user_id')[['revenue', 'test_revenue']].sum().reset_index()
r_pcg1 = pcg1.groupby('user_id')[['revenue', 'test_revenue']].sum().reset_index()
r_pcg2 = pcg2.groupby('user_id')[['revenue', 'test_revenue']].sum().reset_index()

tg = tg.merge(r_ptg, left_on='user_id', right_on='user_id', how='left')
cg1 = cg1.merge(r_pcg1, left_on='user_id', right_on='user_id', how='left')
cg2 = cg2.merge(r_pcg2, left_on='user_id', right_on='user_id', how='left')
cg = pd.concat([cg1,cg2])
```

```python
for df in [tg,cg1,cg2,cg]:
    df.loc[:,['revenue', 'test_revenue']] = df.loc[:,['revenue', 'test_revenue']].fillna(axis=1, value=0)
```

Since statistic is not normal, using non-parametrised test

```python
# aa testing 
n_sample = 3000
samples = []
data = list(tg['revenue'])

for q in range(8):
    random.shuffle(data)
    samples.append(data[:n_sample])
    
i = 0
p_vals = []
for x,y in itertools.permutations(list(range(8)), 2):
    pval = stat.mannwhitneyu(samples[x], samples[y]).pvalue
    if pval < 0.05:
        i+=1
    p_vals.append(pval)
i, p_vals       
```

```python
stat.mannwhitneyu(cg1['test_revenue'], cg2['test_revenue']).pvalue, \
stat.mannwhitneyu(cg1['revenue'], cg2['revenue']).pvalue
```

```python
# Testing by countries 
# since different prices by country are easy to implement
```

```python
ctg = tg.groupby('country')[['user_id']].count().reset_index().rename({'user_id': 'user_id'}, axis='columns')
ccg1 = cg1.groupby('country')[['user_id']].count().reset_index().rename({'user_id': 'user_id_control1'}, axis='columns')
ccg2 = cg2.groupby('country')[['user_id']].count().reset_index().rename({'user_id': 'user_id_control1'}, axis='columns')

cnt = ctg.merge(ccg1, left_on='country', right_on='country').merge(ccg2,left_on='country', right_on='country')
test_countries = list(cnt[cnt['user_id'] > 20]['country'].unique())
cnt
```

```python
test_rez = []

for country in test_countries:
    
    tg_rev = tg[tg['country'] == country][['revenue']]
    cg_rev = cg[cg['country'] == country][['revenue']]
    avg1 = float(tg_rev.sum() / tg_rev.shape[0])
    avg2 = float(cg_rev.sum() / cg_rev.shape[0])
    
    test_rez.append(
        ( country, avg1, avg2, stat.mannwhitneyu(tg_rev, cg_rev).pvalue < 0.05 )
    )
    
test_results = pd.DataFrame(test_rez)
test_results.set_index(0, inplace=True)
test_results.rename({1: 'full target average', 2: 'full control average', 3: 'difference significant' }, axis='columns', inplace=True)
test_results
```

```python
test_rez = []

for country in test_countries:
    
    tg_rev = tg[tg['country'] == country][['test_revenue']]
    cg_rev = cg[cg['country'] == country][['test_revenue']]
    avg1 = float(tg_rev.sum() / tg_rev.shape[0])
    avg2 = float(cg_rev.sum() / cg_rev.shape[0])
    
    test_rez.append(
        ( country, avg1, avg2, stat.mannwhitneyu(tg_rev, cg_rev).pvalue < 0.05 )
    )
test_results1 = pd.DataFrame(test_rez)
test_results1.set_index(0, inplace=True)
test_results1.rename({1: 'subscription target average', 2: 'subscription control average', 3: 'difference significant' }, axis='columns', inplace=True)
```

```python
full_test = test_results.merge(test_results1, left_index=True, right_index=True, how='inner')
full_test
```
